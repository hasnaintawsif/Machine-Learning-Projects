Class 0 = Non-spam (majority class)

Class 1 = Spam (minority class)

This dataset is somewhat imbalanced, but not extremely (roughly 60/40 split).

```python
df.corr().T
```

```python
print('Shape of the Dataset before Selecting Features based on correlation: ', df.shape)
print('\n######### Correlation of each feature with the target class ##########\n')
corr_with_class = df.corr()['class'].drop('class')
print(corr_with_class)
```

Let's select all features with correlation to the target greater than +0.1 or less than -0.1 

```python
selected_features = corr_with_class[corr_with_class.abs() > 0.1].index.tolist()
print(selected_features)
```

Let's marge these Features with class (Target)

```python
df = pd.concat([df[selected_features], df['class']], axis=1)
print('\nShape of the Dataset after Selecting Features based on correlation: ',df.shape)
```
